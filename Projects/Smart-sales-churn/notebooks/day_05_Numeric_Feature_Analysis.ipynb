# Day 5 Task: Numeric Feature Analysis & Visualization

```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Optional: make plots prettier
sns.set(style="whitegrid")

```

```python
# Load cleaned data
df = pd.read_csv(r"../data/cleaned_telco.csv")
```

```python
# Identify numeric and categorical columns

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Exclude target column from categorical columns
if 'churn_label' in categorical_cols:
    categorical_cols.remove('churn_label')

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
df['total_charges'] = df['total_charges'].fillna(0)
df.fillna({'total_charges': 0}, inplace=True)
```

```python
df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```

```python
# List of numeric features to analyze
numeric_cols = ['tenure_months', 'monthly_charges', 'total_charges']

# Check basic stats
df[numeric_cols].describe()
```

```python
# Task 2: Mean values for numeric columns grouped by churn
print(df.groupby('churn_label')[numeric_cols].mean())
```

```python
# Task 3: Boxplot for each numeric feature vs churn
import matplotlib.pyplot as plt
import seaborn as sns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='churn_label', y=col, data=df)
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
# Task 4: Countplot for each categorical feature vs churn
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, hue='churn_label', data=df)
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.show()
```
