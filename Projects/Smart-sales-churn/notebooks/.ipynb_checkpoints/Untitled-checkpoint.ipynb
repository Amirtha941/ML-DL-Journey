```python
#Cell 1 — Import Libraries & Load Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv(r"../data/cleaned_telco.csv")

# Quick info
df.info()
df.head()
```

```python
#Analyze Categorical Features

categorical_cols = ['gender','partner','dependents','internet_service','contract','payment_method']

for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
    df[col].value_counts().plot(kind='bar', title=col)
    plt.show()
```

```python
#Visualize Churn vs Key Features

# Churn distribution for contract type
sns.countplot(data=df, x='contract', hue='churn_label')
plt.title("Churn vs Contract Type")
plt.show()

# Churn distribution for internet service
sns.countplot(data=df, x='internet_service', hue='churn_label')
plt.title("Churn vs Internet Service")
plt.show()
```
