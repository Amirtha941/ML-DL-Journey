## üõ†Ô∏è Step 1: Import Libraries

```python

# üõ†Ô∏è Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (8, 5)


```

## üìÇ Step 2: Loading data

```python
# Loading data
df = pd.read_csv(r"../data/Telco_customer_churn.csv")


# Quick look
print("‚úÖ Dataset Loaded Successfully!")
print("Shape of dataset:", df.shape)
df.head()
```

## üîç Step 3: Basic Info

```python
df.info()
```

```python
df.describe(include="all").T
```

 ## üßπ Step 4: Clean Column Names (optional)

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("‚úÖ Cleaned Column Names:")
print(df.columns.tolist())
```

## Day-2 Data Preprocessing

## üéØ Step 5: Identify Target Column

```python
target_col = 'churn_label'
if target_col not in df.columns:
    print("‚ö†Ô∏è Check target column name ‚Äî possible options:")
    print([col for col in df.columns if 'churn' in col.lower()])
else:
    print(f"‚úÖ Target column found: {target_col}")
```

## üìä Step 6: Target Distribution

```python
if target_col in df.columns:
    print(df[target_col].value_counts())
    sns.countplot(x=target_col, data=df)
    plt.title('Churn Distribution')
    plt.show()
```

## üß© Step 7: Data Type Check & Missing Values

```python
print("\nüîé Missing Values:\n", df.isna().sum())
print("\nüîé Data Types:\n", df.dtypes)

# Example: Convert TotalCharges to numeric if needed
if 'totalcharges' in df.columns:
    df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
    print("\n‚úÖ Converted 'totalcharges' to numeric type.")
    print("Missing values after conversion:\n", df['totalcharges'].isna().sum())
```

## üìò Step 8: Initial Observations

```python
print("üí° Observations to Note:")
print("- Dataset Shape:", df.shape)
print("- Target Variable Distribution checked.")
print("- Missing values handled/identified.")
print("- Column names standardized.")
print("- Ready for Day 2 preprocessing!")
```
